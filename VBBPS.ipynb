### Regression Neural Network From scratch

#### Simple Model

```python
from sklearn.ensemble import GradientBoostingRegressor,RandomForestRegressor,VotingRegressor
from sklearn.model_selection import RandomizedSearchCV
```

```python
m1 = GradientBoostingRegressor(n_estimators=750)
params = [{"learning_rate":[0.1,0.5,1,2],"max_depth":[3,4,5],"random_state":[None,42]}]
```

```python
m1.get_params()
```

```python
modelSimplexRegression = RandomizedSearchCV(m1,params,cv=3,scoring="mean_square_error",return_train_score=True)
```

```python
modelSimplexRegression.estimator
```

```python
import pandas as pd
```

```python
# NUERAL NETWROK
from sklearn.pipeline import Pipeline
```

```python
m2 = VotingRegressor(estimators=[("m1",modelSimplexRegression),("m2",modelSimplexRegression)])
m_2=VotingRegressor(estimators=[("m_1",m2),("m_2",m2)])
m__2=VotingRegressor(estimators=[("m__1",m_2),("m__2",m_2)])
m___2=VotingRegressor(estimators=[("m___1",m__2),("m__2",m__2)])
modelNeuralNetwork = Pipeline([
    ("l1",m2),
    ("l2",m_2),
    ("l3",m__2),
    ("l4",m___2)
])
```

```python
modelComplexRegression = RandomizedSearchCV(modelNeuralNetwork,params,cv=3,scoring="mean_square_error",return_train_score=True)
```

### Classification NeralNetwork from scratch

#### simnplex model

```python
from sklearn.ensemble import AdaBoostClassifier,VotingClassifier,RandomForestClassifier
mc1 = AdaBoostClassifier(estimator=RandomForestClassifier,n_estimators=750)
paramsC = [{"learning_rate":[0.1,0.5,1,2],"n_estimators":[250,500,750],"random_state":[None,42]}]
mc1.get_params()
```

```python
modelsimplexClassification = RandomizedSearchCV(mc1,paramsC,cv=3,scoring="precision",return_train_score=True)
```

### Neural network

```python
m3 = VotingClassifier(estimators=[("m1",modelsimplexClassification),("m2",modelsimplexClassification)])
m_3=VotingClassifier(estimators=[("mo1",m3),("mo2",m3)])
m__3=VotingClassifier(estimators=[("moo1",m_3),("moo2",m_3)])
m___3=VotingClassifier(estimators=[("mooo1",m__3),("mooo2",m__3)])
modelNeuralNetworkClassification = Pipeline([
    ("l1",m3),
    ("l2",m_3),
    ("l3",m__3),
    ("l4",m___3)
])
modelComplexClassification = RandomizedSearchCV(modelNeuralNetworkClassification,paramsC,cv=3,scoring="precision",return_train_score=True)
```
